```python
import numpy as np
import pandas as pd
from scipy import stats
data=[2,3,4,5,6]
mean_value=np.mean(data)
print("Mean:"+str(mean_value))
```

```python
import numpy as np
data=[2,3,4,5,6]
variance=np.var(data, ddof=1)
std_dev=np.std(data, ddof=1)
print(f"variance: {variance},standard Deviation: {std_dev}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data={
'Math':[40,50,60,70,80],
'Science':[50,60,70,80,90],
'English':[78,70,80,90,100],
'Social':[89,79,56,70,67]
}
df=pd.DataFrame(data)
correlation_matrix=df.corr()
plt.figure(figsize=(8,4))
sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm",fmt=".5f")
plt.title("Student Scores Correlation Heatmap")
plt.show()
```

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=[10,50,20,56,80]
df=pd.DataFrame(data,columns=['values'])
Q1=df['values'].quantile(0.25)
Q3=df['values'].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
outliers=df[(df['values']<lower_bound)|(df['values']>upper_bound)]
plt.boxplot(df['values'])
plt.title("Box plot to detect outliers")
plt.show()
print("outliers:\n",outliers)
```

```python
from scipy import stats
import numpy as np
data = [10,12,14,17,20,30,100]
data_array= np.array(data)
z_scores=np.abs(stats.zscore(data_array))
outliers =data_array[z_scores>3]
print("outliers using z-score method:",outliers)
```
